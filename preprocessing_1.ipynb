```python
import os
import librosa
import numpy as np
import pandas as pd
```

```python
target_dir = "datasets/final_audio"

features = []
labels = []

for file in os.listdir(target_dir):
  label = file.split('_')[0]
  file_path = os.path.join(target_dir, file)
  
  with open(file_path, 'rb') as _file:
    audio, sample_rate = librosa.load(_file, sr=22050) # sr*T => 22050 * 30
    feature = np.array([])
    mfccs = np.mean(librosa.feature.mfcc(y=audio, sr=sample_rate, n_mfcc=40).T, axis=0)
    feature = np.hstack((feature, mfccs))
    chroma = np.mean(librosa.feature.chroma_stft(y=audio, sr=sample_rate).T, axis=0)
    feature = np.hstack((feature, chroma))
    mel = np.mean(librosa.feature.melspectrogram(y=audio, sr=sample_rate).T, axis=0)
    feature = np.hstack((feature, mel))
  
  features.append(feature)
  labels.append(label)
```

```python
directory = "datasets/final_audio/"

spectrograms = []
mfccs_list = []
labels = []
max_spec_shape = (0, 0)

for filename in os.listdir(directory):
  file_path = os.path.join(directory, filename)

  y, sr = librosa.load(file_path, sr=None)
  
  # Compute power spectrogram with Short-Time Fourier Transform
  # Sirens usually cover a broad frequency range so higher values of n_fft has been used
  # Smaller hop_length values can help capture rapid changes in the signal, potentially making it easier to detect sirens against noisy backgrounds
  stft = np.abs(librosa.stft(y, n_fft=4096, hop_length=512))
  spectrogram = librosa.amplitude_to_db(stft)

  mfccs = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=13)

  # Update max spectrogram shape
  if spectrogram.shape[0] > max_spec_shape[0]:
    max_spec_shape = (spectrogram.shape[0], max_spec_shape[1])
  if spectrogram.shape[1] > max_spec_shape[1]:
    max_spec_shape = (max_spec_shape[0], spectrogram.shape[1])

  label = filename.split("_")[0]
  
  spectrograms.append(spectrogram)
  mfccs_list.append(mfccs)
  labels.append(label)
```

```python
# Pad or truncate spectrograms to have the same shape
for i in range(len(spectrograms)):
  spec_shape = spectrograms[i].shape
  pad_width = ((0, max_spec_shape[0] - spec_shape[0]), (0, max_spec_shape[1] - spec_shape[1]))
  spectrograms[i] = np.pad(spectrograms[i], pad_width, mode='constant')
```

```python
spectrograms = np.array(spectrograms)
mfccs_list = np.array(mfccs_list)
labels = np.array(labels)
```
